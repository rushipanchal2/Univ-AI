```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Read the speech.txt file and create a word list
file_name = "speech.txt"
word_list = []

with open(file_name,'r') as f:
    line = f.read()
    line=line.replace("–"," ")
    line=line.replace(":"," ")
    line=line.replace(";"," ")
    line=line.replace("."," ")
    line=line.replace(","," ")

    for line in line.split():
        word_list.append(line.lower())

#flat_list = [item for sublist in word_list for item in sublist]

#word_list = flat_list

print(word_list)
print(len(word_list))


    
```

```python
#Selecting stopwords that belong to English language

#stop_words = df[___].Words.___

df = pd.read_csv("stopwords.csv")
stop_words = (df.loc[:,['Words','Language']]) 
new_stop_words = df.loc[df.Language =="English"]
stop_words = list(new_stop_words['Words'])

#[i for df["Words"] where df.Language =="English"]
#df["word"].tolist

print((stop_words))
print(len(stop_words))
```

```python
#Removing stopwords from the word list extracted from the play

pruned_list = [i for i in word_list if i  not in stop_words]
print(pruned_list)
print(len(pruned_list))
```

```python
### edTest(test_count) ###
#Creating a frequency dictionary with words as keys and count of their occurrences as values 

frequency_dict = {}

for i in pruned_list:
    if i  not in frequency_dict:
        frequency_dict[i] = 1
    else:
        frequency_dict[i] += 1
        
print(frequency_dict)        
```
